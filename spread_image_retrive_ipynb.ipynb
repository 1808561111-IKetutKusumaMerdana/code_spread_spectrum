```python
import cv2
import matplotlib
from PIL import Image
import array
import numpy as np
from numpy import *
from random import gauss
from random import seed
```

```python
grap = cv2.imread("test.png")
row_audio=len(grap)
print(row_audio)
print("grap modified")
ta= [ [ 0 for i in range(8) ] for j in range(row_audio) ]
i=0
j=0
while(j<row_audio):
    while(i<8):
       ta[j][i]=grap[j,i,2]
       i+=1
    j+=1
    i=0
print(ta) #pixel values

```

```python
        
#retrieving begins
tb= [ [ 0 for i in range(8) ] for k in range(row_audio) ]
k=0
i=0
while(k<row_audio):
   for i in range(8):
       if i==0:
           ta[k][i]=0*(256-ta[k][i])
       else:
           tb[k][i]=(ta[k][i])/(2**8)
       i+=1   
   k+=1
   i=0
print("decimal=")
print(tb)#converting i values back to decimal value
```

```python

sr= [ [ 0 for i in range(8) ] for j in range(row_audio) ]
key3= '5'
length=len(key3)
p=list();
for i in range(length):
   p.append(0)
w=0   
for v in key3:
   p[w]=int(v)
   w+=1   
w=0
v=0
print(p)#key3

j=0
i=0
l=0
while(j<row_audio):
   for l in p:
      if i>7:
         i=0
      else:
         sr[j][i]=tb[j][l]
      i+=1
      l+=1
   l=p[0]   
   j+=1
   i=0
print("interleaving retrieved")
print(sr)#interleaving done
o=0
p=0
#lsb changed
```
